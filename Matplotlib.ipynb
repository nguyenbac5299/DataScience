#4. Plotting with pandas using the OO method
* When plotting something quickly use the pyplot method
* when plotting something advanced use the OO method

```python
heart_disease.head()
```

```python
over_50= heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
#Pyplot methodd
over_50.plot(x= 'age',
             y='chol',
             c='target',
             kind='scatter'); #c= color
```

```python
# OO method mixed with pyplot method
fig, ax= plt.subplots(figsize= (10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c= 'target',
             ax=ax);
```

```python
fig1, ax1= plt.subplots(figsize= (10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c= 'target',
             ax=ax1);
ax1.set_xlim([45,100])
```

```python
# OO method from scratch
fig, ax= plt.subplots(figsize=(10,6))
#Plot the data
scatter= ax.scatter(x= over_50['age'],
           y= over_50['chol'],
           c= over_50['target'])
#Custumize the plot
ax.set(title='Heart Disease and Cholesterol Level',
       xlabel='Age',
       ylabel='Cholesterol')
#Add Legend
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
over_50.head()
```

```python
#Subplot of chol, age, thalach
fig, (ax0, ax1)= plt.subplots(ncols=1,
                              nrows=2,
                              figsize=(10,6),
                              sharex=True);

scatter0= ax0.scatter(x= over_50['age'],
                      y= over_50['chol'],
                      c= over_50['target'])

ax0.legend(*scatter0.legend_elements(),title='Target')

ax0.set(title='Heart disease and cholesterol levels',
        ylabel='Cholesterol')
ax0.axhline(over_50['chol'].mean(),
            linestyle='--')



scatter1= ax1.scatter(x= over_50['age'],
                      y= over_50['thalach'],
                      c= over_50['target'])

ax1.legend(*scatter1.legend_elements(),title='Target')

ax1.set(title='Heart disease and cholesterol levels',
        xlabel='Age',
        ylabel='Thalach')
ax1.axhline(over_50['thalach'].mean(),
            linestyle='--');

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```

##Changing th cmap

```python
# OO method from scratch
plt.style.use('seaborn-white')

fig, ax= plt.subplots(figsize=(10,6))
#Plot the data
scatter= ax.scatter(x= over_50['age'],
           y= over_50['chol'],
           c= over_50['target'],
           cmap='winter') #this changes the colour scheme
#Custumize the plot
ax.set(title='Heart Disease and Cholesterol Level',
       xlabel='Age',
       ylabel='Cholesterol')
#Add Legend
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

## Changing the xlim & ylim

```python
#Subplot of chol, age, thalach
fig, (ax0, ax1)= plt.subplots(ncols=1,
                              nrows=2,
                              figsize=(10,6),
                              sharex=True);

scatter0= ax0.scatter(x= over_50['age'],
                      y= over_50['chol'],
                      c= over_50['target'],
                      cmap='winter')

ax0.legend(*scatter0.legend_elements(),title='Target')

ax0.set(title='Heart disease and cholesterol levels',
        ylabel='Cholesterol')
ax0.axhline(over_50['chol'].mean(),
            linestyle='--')
# Change the x limit
ax0.set_xlim([50,80])



scatter1= ax1.scatter(x= over_50['age'],
                      y= over_50['thalach'],
                      c= over_50['target'],
                      cmap='winter')

ax1.legend(*scatter1.legend_elements(),title='Target')

ax1.set(title='Heart disease and cholesterol levels',
        xlabel='Age',
        ylabel='Thalach')
ax1.set_xlim([50,80])

ax1.axhline(over_50['thalach'].mean(),
            linestyle='--');

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
